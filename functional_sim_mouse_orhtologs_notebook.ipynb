```python
from pprint import pprint
import os
import sys
import pandas as pd
import urllib.request
from BioLink.biolink_client import BioLinkWrapper
import ontobio
from mygene import MyGeneInfo
from Modules.Mod1A_functional_sim import FunctionalSimilarity
from Modules.ortholog_traversal import OrthologTraversal
```

```python
blw = BioLinkWrapper()


#genes_url = "http://genomics.lbl.gov/~marcin/XYZ_ncbigeneids.txt"
genes_url = "http://genomics.lbl.gov/~marcin/XYZ_HGNCids.txt"

genesetdata = []
print("getting gene set data")
with urllib.request.urlopen(genes_url) as url:
    genesetdata.append(url.read().decode().replace("\"", "").split('\n'))
```

```python
#find NA
findstr = 'NA'
geneset = []
for g in genesetdata[0]:
    row = g.split("\t")
    if row[-1] != "NA" and row[-1] != "":
        geneset.append(row[-1])
```

```python
ortholog_associations = blw.get_orthoglog_gene_set(gene_set=geneset[:10], orth_taxon_name='mouse')
```

```python
mouse_geneset = []
for assoc in ortholog_associations:
    for ortho in assoc['associations']:
        mouse_geneset.append(ortho['object']['id'])



mod1a_input_object = {
    'input': mouse_geneset,
    'parameters': {
        'taxon': 'mouse',
        'threshold': 0,
    },
}

fsim = FunctionalSimilarity()
fsim.load_input_object(mod1a_input_object)
fsim.load_gene_set()
```

```python
fsim.load_associations()
```

```python
mouse_fsim_genes = fsim.compute_similarity(group='mouse')
mouse_fsim_genes_df = pd.DataFrame(data=mouse_fsim_genes)
```
